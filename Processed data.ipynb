```python
numerical_columns = df[['CreditScore', 'Age', 'Tenure', 'Balance', 'EstimatedSalary', 'Satisfaction Score', 'Point Earned']]

# Calculate z-scores
z_scores = numerical_columns.apply(zscore)

# Identify outliers using Z-Score method
outliers_zscore = z_scores.abs() > 3
print("Outliers detected using Z-Score method:")
print(outliers_zscore.sum())
```

```python
columns_with_outliers = ['CreditScore', 'Age', 'Tenure', 'Balance', 'EstimatedSalary', 'Satisfaction Score', 'Point Earned']

# Check if all specified columns exist in the DataFrame
missing_columns = [col for col in columns_with_outliers if col not in df.columns]
if missing_columns:
    print(f"Warning: The following columns are missing from the DataFrame: {missing_columns}")
else:
    # Set the figure size for the plots
    plt.figure(figsize=(15, 8))
    
    # Create subplots for each column
    for i, column in enumerate(columns_with_outliers, 1):
        plt.subplot(2, 4, i)  # Create a subplot for each column
        sns.boxplot(y=df[column])  # Use y=df[column] to plot the boxplot
        plt.title(column)  # Set the title for each subplot

    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()  # Display the plots
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns

# Example dataset (replace with your actual data)
data = pd.DataFrame({
    'CreditScore': [600, 700, 650, 800, 900, 850, 3000],  # Example outlier in CreditScore
    'Age': [25, 30, 35, 40, 45, 50, 100],  # Example outlier in Age
})

# Initialize Isolation Forest
iso_forest = IsolationForest(contamination=0.2, random_state=42)

# Fit the model and predict outliers
data['Outlier'] = iso_forest.fit_predict(data)

# Outlier values: 1 for inliers, -1 for outliers
print(f"Unique values in the 'Outlier' column: {data['Outlier'].unique()}")
print(f"Number of outliers: {sum(data['Outlier'] == -1)}")
print(f"Number of inliers: {sum(data['Outlier'] == 1)}")

# Check shape before removing outliers
print(f"Shape before removing outliers: {data.shape}")

# Remove rows with outliers (Outlier == -1)
data_clean = data[data['Outlier'] == 1]

# Check shape after removing outliers
print(f"Shape after removing outliers: {data_clean.shape}")

# Visualize the data with a scatter plot (Inliers vs Outliers)
plt.scatter(data['CreditScore'], data['Age'], c=data['Outlier'], cmap='coolwarm', edgecolor='k')
plt.title("Inliers vs Outliers")
plt.xlabel("CreditScore")
plt.ylabel("Age")
plt.show()

# Optionally, visualize the cleaned data
plt.scatter(data_clean['CreditScore'], data_clean['Age'], color='blue', edgecolor='k')
plt.title("Cleaned Data (No Outliers)")
plt.xlabel("CreditScore")
plt.ylabel("Age")
plt.show()
```
